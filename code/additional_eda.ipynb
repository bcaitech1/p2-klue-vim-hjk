```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl
import pprint

from IPython.display import display


fontpath = '/usr/share/fonts/NanumGothicCoding.ttf'
font = fm.FontProperties(fname=fontpath, size=9)

plt.rc('font', family='NanumGothicCoding')
```

```python
column_names = ['information', 'sentence', 'entity_1', 'start_1', 'end_1', 'entity_2', 'start_2', 'end_2', 'relation']
df = pd.read_csv('../input/data/train/train.tsv', sep='\t', names=column_names)
display(df.head())
```

```python
df_relation_counts = df.relation.value_counts().reset_index()
df_relation_counts.columns = ['relation', 'counts']
display(df_relation_counts)
fig, ax = plt.subplots(figsize=(70, 25))
sns.barplot(ax=ax, data=df_relation_counts, x='relation', y='counts')
plt.show()
```

```python
df_2 = pd.read_csv('../input/data/train/train+all.tsv', sep='\t', names=column_names)
df_relation_counts = df_2.relation.value_counts().reset_index()
df_relation_counts.columns = ['relation', 'counts']
display(df_relation_counts)
fig, ax = plt.subplots(figsize=(70, 25))
sns.barplot(ax=ax, data=df_relation_counts, x='relation', y='counts')
plt.show()
```

```python
num_nothing = df_relation_counts['counts'][df_relation_counts.relation.str.contains('관계')].sum()
num_organization = df_relation_counts['counts'][df_relation_counts.relation.str.contains('단체:')].sum()
num_character = df_relation_counts['counts'][df_relation_counts.relation.str.contains('인물:')].sum()

main_cat_df = pd.DataFrame(columns=['main_category', 'counts'])
main_cat_df.loc[0] = ['관계_없음', num_nothing]
main_cat_df.loc[1] = ['단체', num_organization]
main_cat_df.loc[2] = ['인물', num_character]

display(main_cat_df)

fig, ax = plt.subplots(figsize=(3, ))
sns.barplot(ax=ax, data=main_cat_df, x='main_category', y='counts')
plt.show()
```
